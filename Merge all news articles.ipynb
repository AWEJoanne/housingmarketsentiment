```python
import pandas as pd
```

```python
# read scraped data
cna = pd.read_excel('Output/CNA/cleaned_cna_articles_2020_Jan_to_2025_Oct.xlsx')
st = pd.read_excel('Output/ST/cleaned_st_articles_2020_Jan_to_2025_Oct.xlsx')
bt = pd.read_excel('Output/BT/cleaned_bt_articles_2020-01-01_to_2025-10-31.xlsx')

# label the news site
cna['news_site'] = 'cna'
st['news_site'] = 'st'
bt['news_site']= 'bt'
```

```python
cna.info()
```

```python
st.info()
```

```python
bt.info()
```

```python
# merge the data
df = pd.concat([cna, st, bt], ignore_index=True)
```

```python
df.info()
```

```python
# clean the articles by removing newline characters and strip extra spaces

df['content'] = (
    df['content']
    .astype(str)
    .str.replace(r'\s*\n\s*', ' ', regex=True)
    .str.replace(r'\s+', ' ', regex=True)
    .str.strip()
)
```

```python
# recompute the number of content chars

df['content_chars'] = df['content'].astype(str).str.len()
```

```python
df.to_excel('Output/cleaned_news_articles_2020_Jan_to_2025_Oct.xlsx', index=False)
```
